# - Model test : 

## 1- Packages : 

```python
import os
import cv2
import argparse
import numpy as np
#drive.mount('/content/drive')
from google.colab import drive
from google.colab.patches import cv2_imshow
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.image import img_to_array
```

## 2- Loading model and image for testing : 

```python
dirpath = '/content/drive/My Drive'

# Chemin du modèle
modelPath = 'models/system.h5'

# Vérifiez si le fichier existe
if not os.path.exists(modelPath):
    raise FileNotFoundError(f"No file or directory found at {modelPath}")

# Chargement du modèle de détecteur de pneumonie
print("Chargement du modèle de détecteur de pneumonie...")
model = load_model(modelPath)

# Chemin de l'image
imagePath = os.path.join(dirpath, 'chest_xray/test/NORMAL/IM-0003-0001.jpeg')

# Vérifiez si l'image existe
if not os.path.exists(imagePath):
    raise FileNotFoundError(f"No file or directory found at {imagePath}")

# Chargement de l'image
image = cv2.imread(imagePath)
if image is None:
    raise ValueError(f"Could not read the image from {imagePath}")
```

## 3- Image preprocessing : 

```python
# Prétraitement de l'image
image = cv2.resize(image, (150, 150))  # redimensionner à 150x150 pixels
image = image.astype("float") / 255.0  # normaliser les valeurs des pixels
image = img_to_array(image)
image = np.expand_dims(image, axis=0)
```

## 4- prédiction sur l'image : 

```python
# Prédiction
print("Classification de l'image...")
predictions = model.predict(image)
print(f"Sortie du modèle: {predictions}")
```

## 5- Interprétation de la sortie et affichage de l'image : 

```python
# Interprétation de la sortie
if len(predictions[0]) == 2:
    (normal, pneumonia) = predictions[0]
    print(f"Normal: {normal}, Pneumonia: {pneumonia}")
    if pneumonia > normal:
        print("Pneumonia detected.")
    else:
        print("Normal.")
else:
    probability = predictions[0][0]
    print(f"Probabilité de pneumonie: {probability}")
    if probability > 0.5:
        print("Pneumonia detected.")
    else:
        print("Normal.")
```

```python
# Affichage de l'image
cv2_imshow(cv2.imread(imagePath))
if len(predictions[0]) == 2:
    (normal, pneumonia) = predictions[0]
    print(f"Normal: {normal}, Pneumonia: {pneumonia}")
    if pneumonia > normal:
        print("Pneumonia detected.")
    else:
        print("Normal.")
else:
    probability = predictions[0][0]
    print(f"Probabilité de pneumonie: {probability}")
    if probability > 0.5:
        print("Pneumonia detected.")
    else:
        print("Normal.")
```
